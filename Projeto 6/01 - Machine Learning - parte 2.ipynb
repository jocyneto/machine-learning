```python
import pandas as pd
```

```python
url = 'https://drive.google.com/file/d/17pvE1IiR-nC0w3cZNfikR75uhSbNFd8a/view?usp=sharing'
url2= 'https://drive.google.com/uc?id=' + url.split('/')[-2]
dados = pd.read_csv(url2)
dados.head()
```

```python
dados = pd.get_dummies(dados)
dados.head()
```

```python
x = dados[['home','logado','busca_algoritmos','busca_java','busca_ruby']]
y = dados['comprou']
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

SEED = 5
np.random.seed(SEED)

treino_x,teste_x,treino_y,teste_y = train_test_split(x, y, test_size = 0.3, stratify =y)

modelo = MultinomialNB()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)
acuracia = accuracy_score(teste_y, previsoes) * 100
print(f'acuracia: {acuracia:.2f}')
```

```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier(strategy = 'most_frequent')
dummy.fit(treino_x, treino_y)
dummy.predict(teste_x)
acuracia_dummy = dummy.score(teste_x, teste_y) * 100
print(f'acuracia: {acuracia_dummy:.2f}')
```
