```python
import pandas as pd

uri = 'https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv'
dados = pd.read_csv(uri)
```

```python
dados.head()
```

```python
map_col = {
    'home':'principal',
    'how_it_works':'como_funciona',
    'contact':'contato',
    'bought':'comprou'
} 
dados.rename(columns = map_col, inplace = True)
```

```python
x = dados[['principal', 'como_funciona','contato']]
```

```python
y = dados[['comprou']]
```

```python
# Pegar 75% para treino
# Pegar 25% para teste
tamanho_treino = round(len(x) * 0.75)

treino_x = x[:tamanho_treino]
treino_y = y[:tamanho_treino]
teste_x = x[tamanho_treino:]
teste_y = y[tamanho_treino:] # Os dados reais

print(f'Neste modelo será treinado com {len(treino_x)} registros e testado com {len(teste_x)} registros.')
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

model = LinearSVC()
model.fit(treino_x, treino_y)

previsoes = model.predict(teste_x)

taxa_de_acerto = accuracy_score(teste_y, previsoes)*100

print(f'Taxa de acerto do modelo: {taxa_de_acerto:.2f}%')
```

# Usado separador de treino e teste automaticamente

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
SEED = 20 #Numero para fixar a escolha do numero aleatorio, assim, não demonstra varios resultados 
          #diferentes a cada vez que rodar o algoritmo, pois irá tirar a aleatoridade da seleção dos dados de teste e treino

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, random_state = SEED, test_size = 0.25, stratify = y)
                                                                                                        # separa proporcionalmente o treino e teste
print(f'Neste modelo será treinado com {len(treino_x)} registros e testado com {len(teste_x)} registros.')

model = LinearSVC()

model.fit(treino_x, treino_y)
previsoes = model.predict(teste_x)
taxa_acerto = accuracy_score(teste_y, previsoes)*100
print(f'Taxa de acerto do modelo: {taxa_acerto:.2f}%')
```
