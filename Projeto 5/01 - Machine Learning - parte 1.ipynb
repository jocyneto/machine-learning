```python
#Features (Caracteristicas)    #0=> não / 1=> sim
# pelo longo?
# perna curta?
# faz auau?

porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1] 
```

```python
# 0=> Cachorro / 1=> Porco
treino_x = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
treino_y = [1, 1, 1, 0, 0, 0]
```

```python
from sklearn.svm import LinearSVC
```

```python
# instancia meu modelo de classificacao
model = LinearSVC()
```

```python
# treinar o modelo (fit)
# X = Featurues, suas caracteristicas
# Y = Classes, se é cachorro ou porco, por exemplo.
model.fit(treino_x, treino_y)
```

```python
# predizer o modelo (predict)
# Neste passo é aonde o modelo vai poder predizer para gente com base no treinamento anterior
# , se ele é cachorro ou porco.
# Neste teste ele vai retorna array([0]), dizendo que o que foi passado foi um cachorro
animal_misterioso = [1, 1, 1]
model.predict([animal_misterioso])
                # precisa passar um array de itens, mesmo sendo somente 1 items
```

```python
# Realizando mais predições:
animal_misterioso1 = [1, 1, 1]
animal_misterioso2 = [0, 1, 0]
animal_misterioso3 = [0, 1, 1]

teste_x = [animal_misterioso1, animal_misterioso2, animal_misterioso3]
teste_y = [0, 1, 1]

model.predict(teste_x)
```

```python
# previsao de acerto e erro:
# Para começar preciso pegar a previsao que foi feita pelo modelo
previsao = model.predict(teste_x)

# Depois pegar os valores reais, o que é esperado:
teste_y = [0, 1, 1]

# Apos isso comparamos as previsões do modelo com os valores reais
print(previsao == teste_y,'\n')

# A taxa de acerto poderiamos fazer na mao pegando o numero de acertos do algoritmo rodado no anterior
# Depois pegar o total dos itens e fazer a divisao do total acertos pelo total daria minha previsao
# No ScikitLearn já possui isto implementado utilizando o accuracy_score 
from sklearn.metrics import accuracy_score

taxa_de_acerto = accuracy_score(teste_y, previsao)
print("Taxa de acerto: ", taxa_de_acerto.round(2) * 100)
```
